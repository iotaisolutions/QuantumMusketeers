```python
# This is an Example Circuit.. Here Your circuit will come in

#n_qubits = 4
#circ = QuantumCircuit(n_qubits, n_qubits)

# Test Circuit
#circ.h(0)
#for qubit in range(n_qubits - 1):
#    circ.cx(qubit, qubit + 1)
#circ.measure(range(4), range(4))
#print(circ)

from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister
from qiskit import execute, Aer ,transpile
from qiskit_textbook.tools import array_to_latex
from qiskit.visualization import plot_histogram,plot_bloch_multivector
from qiskit.quantum_info import Statevector 
from qiskit.providers.aer import QasmSimulator
from qiskit.quantum_info import state_fidelity
from qiskit.aqua.algorithms import NumPyLSsolver
import numpy as np
import time

start_time = time.time()

%matplotlib inline

t = np.pi*3/32

#Tridiagonal Matrix Ax=b

theta = (np.pi)/2  # Angle defining |b>

b = 1  # Matrix Lower off-diagonal
a = -1  # Matrix diagonal
c= 1  # Matrix Upper off-diagonal

# Initialise the quantum and classical registers
qrb = QuantumRegister(6) # Input register
qrl = QuantumRegister(6) # Clock Register
qra = QuantumRegister(1) # Ancilla Register
qrclassic = ClassicalRegister(6) #Classic Register for measurement

# Create a Quantum Circuit
qc = QuantumCircuit(qrb,qrl,qra,qrclassic)


# State preparation for Input Registers 
qc.ry(theta, qrb[5])
    
#Hadmard Gates on Clock Registers 
for qu in qrl:
    qc.h(qu)

# # Applying Unitary Operator

for i in range(5,0,-1):
    qc.p(a*t*pow(2,i), qrl[i])
    qc.u(b*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.u(c*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])



for i in range(5,0,-1):
    params=1
    for j in range(5,0,-1):
        params=b*t*c*pow(2,j)
        qc.p(np.pi/2,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(params,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(-params,qrb[i])
        qc.p(3*np.pi/2,qrb[i])
    

# Inverse QFT

for i in range(5,0,-1):
    if i != 0:
        qc.h(qrl[i])
        qc.rz(-np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(-np.pi/4,qrl[i-1])
        qc.h(qrl[i-1])
    else: 
        qc.rz(-np.pi/4,qrl[i])
        qc.h(qrl[i])

# Eigenvalue rotation
t1=(-np.pi +np.pi/3 - 2*np.arcsin(1/3))/4
t2=(-np.pi -np.pi/3 + 2*np.arcsin(1/3))/4
t3=(np.pi -np.pi/3 - 2*np.arcsin(1/3))/4
t4=(np.pi +np.pi/3 + 2*np.arcsin(1/3))/4


for i in range(5,0,-1):
    qc.cx(qrl[i],qra[0])
    qc.ry(t1,qra[0])
    qc.cx(qrl[i-1],qra[0])
    qc.ry(t2,qra[0])
    qc.cx(qrl[i],qra[0])
    qc.ry(t3,qra[0])
    qc.cx(qrl[i-1],qra[0])
    qc.ry(t4,qra[0])

# QFT

for i in range(6):
    if i != 0:
        qc.h(qrl[i])
        qc.rz(-np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(-np.pi/4,qrl[i-1])
        qc.h(qrl[i-1])
    else: 
        qc.rz(-np.pi/4,qrl[i])
        qc.h(qrl[i])
        
# Inverse of Unitary Operator
for i in range(6):
    qc.u(c*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.u(b*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.p(a*t*pow(2,i), qrl[i])




for i in range(6):
    params=1
    for j in range(6):
        params=b*t*c*pow(2,j)
        qc.p(np.pi/2,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(params,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(-params,qrb[i])
        qc.p(3*np.pi/2,qrb[i])        

# Inverse Hadmard Gates on Clock Registers
for qu in qrl:
    qc.h(qu)

#Transpilation for Circuit Optimization (to improve qubit connectivity and reduce depth)
from qiskit.compiler import transpile
circ = transpile(qc,backend=Aer.get_backend('qasm_simulator'),optimization_level=3)

print("Depth: %i" % circ.depth())
gates=circ.count_ops()
circuit_cost = gates['u'] + gates['u1']+ gates['p']+10*gates['cx']
print("Circuit Cost: %i" % circuit_cost)
#circ.measure(qrb,qrclassic)
circ.draw(fold=100,output='mpl')
```
