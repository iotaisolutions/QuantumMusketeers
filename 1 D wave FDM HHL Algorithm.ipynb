```python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister
from qiskit import execute, Aer ,transpile
from qiskit_textbook.tools import array_to_latex
from qiskit.visualization import plot_histogram,plot_bloch_multivector
from qiskit.quantum_info import Statevector 
import numpy as np
import time

start_time = time.time()

%matplotlib inline

t = np.pi*3/32

#Tridiagonal Matrix Ax=b

theta = (np.pi)/2  # Angle defining |b>

b = 1  # Matrix Lower off-diagonal
a = -1  # Matrix diagonal
c= 1  # Matrix Upper off-diagonal

# Initialise the quantum and classical registers
qrb = QuantumRegister(6) # Input register
qrl = QuantumRegister(6) # Clock Register
qra = QuantumRegister(1) # Ancilla Register
qrclassic = ClassicalRegister(7) #Classic Register for measurement

# Create a Quantum Circuit
qc = QuantumCircuit(qrb,qrl,qra,qrclassic)


# State preparation for Input Registers 
qc.ry(theta, qrb[5])
    
#Hadmard Gates on Clock Registers 
for qu in qrl:
    qc.h(qu)

# # Applying Unitary Operator

for i in range(5,0,-1):
    qc.u1(a*t*pow(2,i), qrl[i])
    qc.u3(b*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.u3(c*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])



for i in range(5,0,-1):
    params=1
    for j in range(5,0,-1):
        params=b*c*t*pow(2,j)
        qc.u1(np.pi/2,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(params,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(-params,qrb[i])
        qc.u1(3*np.pi/2,qrb[i])
    

# Inverse QFT

for i in range(5,0,-1):
    if i != 0:
        qc.h(qrl[i])
        qc.rz(-np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(-np.pi/4,qrl[i-1])
        qc.h(qrl[i-1])
    else: 
        qc.rz(-np.pi/4,qrl[i])
        qc.h(qrl[i])

# Eigenvalue rotation
t1=(-np.pi +np.pi/3 - 2*np.arcsin(1/3))/4
t2=(-np.pi -np.pi/3 + 2*np.arcsin(1/3))/4
t3=(np.pi -np.pi/3 - 2*np.arcsin(1/3))/4
t4=(np.pi +np.pi/3 + 2*np.arcsin(1/3))/4


for i in range(5,0,-1):
    qc.cx(qrl[i],qra[0])
    qc.ry(t1,qra[0])
    qc.cx(qrl[i-1],qra[0])
    qc.ry(t2,qra[0])
    qc.cx(qrl[i],qra[0])
    qc.ry(t3,qra[0])
    qc.cx(qrl[i-1],qra[0])
    qc.ry(t4,qra[0])

# QFT

for i in range(6):
    if i != 0:
        qc.h(qrl[i])
        qc.rz(-np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(np.pi/4,qrl[i])
        qc.cx(qrl[i-1],qrl[i])
        qc.rz(-np.pi/4,qrl[i-1])
        qc.h(qrl[i-1])
    else: 
        qc.rz(-np.pi/4,qrl[i])
        qc.h(qrl[i])
        
# Inverse of Unitary Operator
for i in range(6):
    qc.u3(c*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.u3(b*t*pow(2,i), -np.pi/2, np.pi/2, qrb[i])
    qc.u1(a*t*pow(2,i), qrl[i])




for i in range(6):
    params=1
    for j in range(6):
        params=b*c*t*pow(2,j)
        qc.u1(np.pi/2,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(params,qrb[i])
        qc.cx(qrl[j],qrb[i])
        qc.ry(-params,qrb[i])
        qc.u1(3*np.pi/2,qrb[i])        

# Inverse Hadmard Gates on Clock Registers
for qu in qrl:
    qc.h(qu)
    

#Transpilation for Circuit Optimization (to improve qubit connectivity and reduce depth)

transpiled_circuit = transpile(qc,backend=Aer.get_backend('qasm_simulator'),optimization_level=2)


print("Depth: %i" % transpiled_circuit.depth())
print("CNOTS: %i" % transpiled_circuit.count_ops()['cx'])
transpiled_circuit.draw(fold=100,output='mpl')
```

```python
#Simulating the quantum circuit

statevector_simulator = Aer.get_backend('statevector_simulator')
sim_result = execute(transpiled_circuit, backend=statevector_simulator, shots=1024).result()
array_to_latex(sim_result.get_statevector(),pretext="\\text {StateVector}= \n")
print("Quantum Algorithm Execution Time Taken--- %s seconds ---" % (time.time() - start_time))
```
